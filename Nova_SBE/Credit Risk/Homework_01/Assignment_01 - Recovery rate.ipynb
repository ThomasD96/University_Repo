```python
import numpy as np
from scipy.stats import beta
from scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# estimate parameters of beta dist
def getAlphaBeta(mu, sigma):
    alpha = mu**2 * ((1 - mu) / sigma**2 - 1 / mu)
    beta = alpha * (1 / mu - 1)
    return {"alpha": alpha, "beta": beta}
```

```python
mydict = getAlphaBeta(0.35, 0.22)
mykeys = ['alpha', 'beta']
alpha_beta =[mydict[k] for k in mykeys if k in mydict]
a = alpha_beta[0]
b = alpha_beta[1]
```

```python
# Generate random values x between 0 and 1
x = np.arange(start=0, stop=1, step=0.001)

# Plot the beta distribution
plt.figure(figsize=(9,5))
plt.xlim(-0.005, 1.001)
plt.ylim(-0.005, 1.80)
plt.plot(x, beta.pdf(x, a, b), '-', color = 'black')
plt.xlabel('Values of Random Variable x (0, 1)', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.savefig('beta_distribution.png', dpi = 100)
plt.show()
```

```python
# make 10,000 random draws from beta distribution
random_draws = np.random.beta(a, b, size=10000)
```
