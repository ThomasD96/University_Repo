```python
import numpy as np
from scipy.stats import beta
from scipy import stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# estimate parameters of beta dist
def getAlphaBeta(mu, sigma):
    alpha = mu**2 * ((1 - mu) / sigma**2 - 1 / mu)
    beta = alpha * (1 / mu - 1)
    return {"alpha": alpha, "beta": beta}
```

```python
mydict = getAlphaBeta(0.35, 0.22)
mykeys = ['alpha', 'beta']
alpha_beta =[mydict[k] for k in mykeys if k in mydict]
a = alpha_beta[0]
b = alpha_beta[1]
```

```python
# Generate the values between
x = np.linspace(beta.ppf(0.001, a, b),beta.ppf(0.999, a, b), 1000)

# Plot the beta distribution
plt.figure(figsize=(9,5))
plt.xlim(0, 1)
plt.plot(x, beta.pdf(x, a, b), '-', color = 'black')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.savefig('beta_distribution.png', dpi = 100)
plt.show()
```
